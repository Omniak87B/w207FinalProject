```python
import pandas as pd
import numpy as np
import os
```

```python
train = pd.DataFrame.from_csv('train.csv')
```

```python
train.head()
```

```python
def five_num_summary(df, column):
    print(f'''
    Column: {column}
    Max value: {np.max(df[column])}
    Min value: {np.min(df[column])}
    Mean: {np.mean(df[column])}
    Median: {np.median(df[column])}
    ''')
for col_name in train.columns:
    five_num_summary(train, col_name)
```
